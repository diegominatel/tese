```python
''' Importa as bibliotecas necessárias '''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
''' Configurações '''
sns.set(rc={'figure.figsize':(7,4)})
sns.set(style="ticks", rc={"grid.linewidth": 0.2})
```

# Experimento 1

```python
# carrega os resultados do experimento 1
exp1 = pd.read_csv('exp1.csv', sep=';')
# troca o nome de Rasch para 1PL (para ficar igual a tese)
exp1 = exp1.replace('Rasch', 'ML1')
exp1 = exp1.replace('2PL', 'ML2')
exp1 = exp1.replace('3PL', 'ML3')
# tira a média dos resultados para as 10 repetições
exp1 = exp1.groupby(['TRI', 'N_ID', 'N_ITENS']).mean().reset_index()
# deixa o valor do tempo relativo ao total (em porcentagem)
exp1['TR'] = 100*exp1['TIME']/exp1['TIME'].sum()
```

```python
'''Gera o gráfico relativo ao Experimento 1'''

# define o label dos eixos
plt.xlabel("Número de Itens")
plt.ylabel("Tempo de Processamento")

# imprime o gráfico de linha
ax = sns.lineplot(data=exp1, x='N_ITENS', y='TR', hue='TRI', palette='husl', linewidth=1.5)

# deixa o eixo y em probabilidades
ticks = ['', '0,00%', '0,25%', '0,50%', '0,75%', '1,00%', '1,25%', '1,50%', '1,75%', '2,00%']
ax.set_yticklabels(ticks)
# corta o gráfico (para deixar padronizado com todos os gráficos da tese)
sns.despine(trim=False, offset=10)
# tira o título da legenda 
ax.legend_.set_title(None)
# tira a borda da legenda
leg = plt.legend()
leg.get_frame().set_facecolor('none')
leg.get_frame().set_linewidth(0.0)

figure = ax.get_figure()
figure.savefig('exp1.pdf', bbox_inches='tight')
```
